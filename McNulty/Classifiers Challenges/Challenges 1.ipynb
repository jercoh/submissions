Import data
---

```python
import pandas as pd
import numpy as np

votes_data_file = open('house-votes-84.data', 'r')
df = pd.DataFrame.from_csv(votes_data_file)
df = df.reset_index()
```

Prepare data
---

```python
# replace y by 1 n by 0
df = df.replace('y', 1)
df = df.replace('n', 0)

# replace ? by the mean of each column
df = df.replace('?', np.nan)
for column, series in df.iteritems():
    if column != 'party':
        mean = df[column].mean()
        df[column] = df[column].replace(np.nan, mean)
```

```python
df.describe()
```

Split data
---

```python
from sklearn.cross_validation import train_test_split

df_training, df_test = train_test_split(df, test_size = 0.30)
```

```python
len(df_training)+len(df_test)
```

```python
df_training[:,0]
```

Classifiers
===

kNN
---

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

for k in range(1,21):
    neighbors = KNeighborsClassifier(n_neighbors=k)
    neighbors.fit(df_training[:,1:], df_training[:, 0])
    predictions = neighbors.predict(df_test[:,1:])
    print 'k = '+str(k)+':'
    print accuracy_score(df_test[:, 0], predictions)
```

Logistic regression
---

```python
from sklearn.linear_model import LogisticRegression

logistic = LogisticRegression()
logistic.fit(df_training[:,1:], df_training[:, 0])
logistic_predictions = logistic.predict(df_test[:,1:])
print 'Logistic Regression:'
print accuracy_score(df_test[:, 0], logistic_predictions)
```

```python
import pickle
fig = plt.figure(figsize=(35, 14))

picklefile = pickle.load(open('movies_dataset.pkl', 'r'))
df = pd.DataFrame(picklefile)

int_columns = ['Metascore', 'imdbVotes', 'tomatoUserMeter', 'tomatoReviews', 'tomatoUserReviews', 'tomatoMeter', 'tomatoFresh']
float_columns = ['imdbRating', 'tomatoRating', 'tomatoUserRating']

# Convert int to en-format
df[int_columns] = df[int_columns].applymap(lambda x: x.replace(',',''))

# Convert Runtime in integer
df['Runtime'] = df.Runtime.map(lambda x: int(str(x).split()[0]) if x != 'N/A' else x)

# replace string by floats and ints
df = df.replace('N/A', '0')
df[float_columns] = df[float_columns].astype(float)
df[['Year']+int_columns] = df[['Year']+int_columns].astype(int)

# replace zeros by np.nan
df = df.replace('0', np.nan)



# replace True by 1, False by 0
df = df.applymap(lambda x: 1 if x == True else (0 if x == False else x))

# manually add the intercept
df['Intercept'] = 1.0
```

```python
import logisticreg_helper as helper

log_reg = LogisticRegression()
log_reg.fit(df[df.columns[:3]], df.survival_status)

# Create a new data sets with all possible values
features = df.columns[:3]
new_values = []
for feature in features:
    values = np.linspace(df[feature].min(), df[feature].max(), 10)
    new_values.append(values)

# enumerate all possibilities
combos = pd.DataFrame(helper.cartesian(new_values))

# recreate the columns
combos.columns = features

# make predictions on the enumerated dataset
combos['survival_status'] = log_reg.predict(combos[features])
```

```python
def isolate_and_plot(columns, data):
    i=0
    colors = 'rbgyrbgy'
    for column in columns:
        axes = plt.gca()
        axes.set_ylim([0,1])
        grouped = pd.pivot_table(data, values=['survival_status'], rows=[column],
                                 aggfunc=np.mean)
        plt.plot(grouped.index.get_level_values(0), grouped['survival_status'],
                    color=colors[int(i)])
        plt.xlabel(column)
        plt.ylabel("P(survived=1)")
        plt.legend(column, loc='upper left', title='Variable')
        plt.title("Prob(survived=1) isolating " + column)
        plt.show()
        i+=1
```

```python
isolate_and_plot(features, combos)
```
